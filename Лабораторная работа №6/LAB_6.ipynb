```python
def distance_line(A,B,M,N):
    AB = Point(B.x-A.x,B.y-A.y,B.z-A.z)
    l1 = Line(A,B)
    l2 = Line(M,N)
    p = Plane(A, normal_vector=AB)
    pl2 = p.projection_line(l2)
    d = pl2.distance(A)
    return(d)
```

```python
def Point_oneside_P(A,B,P):
    s = Segment(A,B)
    p = P.intersection(s)
    if len(p) == 0:
        return True
    else:
        return not(s.contains(p[0]))
```

```python
def Point_opposite_l(A,l):
    A0 = l.projection(A)
    x = 2*A0.x - A.x
    y = 2*A0.y - A.y
    if len(A) == 2:
        return Point(x,y)
    elif len(A) == 3:
        z = 2*A0.z - A.z
        return Point(x,y,z)
```

### Приведение кривой второго порядка к каноническому виду

```python
x = Symbol('x')
y = Symbol('y')

M = Matrix([[x**2,x*y,y**2,x,y,1],
            [1**2,1*1,1**2,1,1,1],
            [0**2,0*1,1**2,0,1,1],
            [1**2,1*0,0**2,1,0,1],
            [2**2,2*3,3**2,2,3,1],
            [3**2,3*2,2**2,3,2,1]])

det(M)
```

```python
def conic_curve(A,a,f_transform=0):
    if (A.shape != (2,2)) or (len(a) != 3):
        raise ValueError('Invalid size of Aya matrices')
    
    a11 = A[0,0]; a12 = A[0,1]; a22 = A[1,1]
    a1 = a[0]; a2 = a[1]; a0 = a[2]
    D = det(A)
    Delta = det(Matrix([[a11,a12,a1],
    [a12,a22,a2],
    [a1, a2, a0]]))

    I = a11+a22
    B = det(Matrix([[a11,a1],
                    [a1, a0]])) + \
        det(Matrix([[a22,a2],
                    [a2, a0]]))
    
    if (Delta*I < 0) and (D > 0):
        print('ellipse')
    if (Delta != 0) and (D < 0):
        print('Hyperbola')
    if (Delta != 0) and (D == 0):
        print('Parabola')
    if (Delta == 0) and (D < 0):
        print(' pair of intersecting spindles')
    if (Delta == 0) and (D == 0) and (c < 0):
        print(' pair of parallel strands')
    if (Delta == 0) and (D == 0) and (B == 0):
        print(' Poyai')
    if (Delta == 0) and (D > e) and (B == e):
        print(' Point')
    if (Delta*I > 0) and (D > e):
        print(' min ellipse')
    if (Delta == 0) and (D == 0) and (B > e):
        print(' pair of imaginary parallel lines')
    
    T, _ = A.diagonalize()
    T1 = T.inv()
    n1 = sqrt(T1[0,0]**2+T1[1,0]**2)
    n2 = sqrt(T1[0,1]**2+T1[1,1]**2)
    x,y,x1,y1 = symbols('x y x1 y1')
    
    Q0 = a11*x**2 + 2*a12*x*y + a22*y**2 + \
        2*a1*x + 2*a2*y + a0
    x0 = (T1[0,0]/n1)*x1+(T1[1,0]/n1)*y1
    y0 = (T1[0,1]/n2)*x1+(T1[1,1]/n2)*y1
    
    Q = Q0.subs({x: x0, y: y0}).simplify()
    if (f_transform == 0):
        print('Equation: %s' % Q)
    else:
        print('Equation: %s' % Q)
        print('transition equation:')
        print('x = %s' % x0)
        print('y = %s' % y0)
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')

coefs = (1, 2, 2)

rx, ry, rz = 1/np.sqrt(coefs)

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

x = rx * np.outer(np.cos(u), np.sin(v))
y = ry * np.outer(np.sin(u), np.sin(v))
z = rz * np.outer(np.ones_like(u), np.cos(v))

ax.plot_surface(x, y, z, rstride=4, cstride=4, color='g')

ax.set_xlabel("$X$")
ax.set_ylabel("$Y$")
ax.set_zlabel("$Z$")

plt.show()
```

# Аналитическое решение.

```python
from matplotlib import pyplot
import numpy
```

Уравнение Лапласа с перечисленными выше граничными условиями имеет аналитическое решение, заданное формулой

\begin{equation}
p(x,y) = \frac{\sinh \left( \frac{\frac{3}{2} \pi y}{L}\right)}{\sinh \left(  \frac{\frac{3}{2} \pi H}{L}\right)} \sin \left( \frac{\frac{3}{2} \pi x}{L} \right)
\end{equation}

где  $L$ и $H$ длина домена в направлениях $x$ и $y$ соответственно.

Мы будем использовать [`numpy.meshgrid`](http://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html) для построения наших 2D-решений. Это функция, которая принимает два вектора (`x` и `y`, сказать) и возвращает 2D массива из $x$ и $y$ координаты, которые мы затем используем для создания контурного графика. Всегда полезно [`linspace`](https://docs.scipy.org/doc/numpy-1.10.0/reference/generated/numpy.linspace.html)создает массивы из 1 строки чисел с равными интервалами: это помогает определить $x$ и $y$ оси на линейных графиках, но теперь мы хотим, чтобы аналитическое решение было построено для каждой точки в нашей области. Чтобы сделать это, мы будем использовать в аналитическом решении двумерные массивы, сгенерированные с помощью `numpy.meshgrid`.

```python
def p_analytical(x, y):
    X, Y = numpy.meshgrid(x,y)
    
    p_an = numpy.sinh(1.5*numpy.pi*Y / x[-1]) /\
    (numpy.sinh(1.5*numpy.pi*y[-1]/x[-1]))*numpy.sin(1.5*numpy.pi*X/x[-1])
    
    return p_an
```

Хорошо, давайте опробуем аналитическое решение и используем его для тестирования функции `plot_3D`, которую мы написали выше.

```python
nx = 41
ny = 41

x = numpy.linspace(0,1,nx)
y = numpy.linspace(0,1,ny)

p_an = p_analytical(x,y)
```

```python
plot_3D(x,y,p_an)
```

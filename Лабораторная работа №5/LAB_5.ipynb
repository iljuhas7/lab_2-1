```python
def Lin_homogen_2(a,yl):
    x = symbols('x')
    u = Function('u')
    z = Function('z')

    y1d = diff(y1,x)

    eq = y1*diff(u(x),x)+(2*y1d+a*y1)*u(x)
    u0 = dsolve(eq,u(x))

    eq = diff(z(x),x)-u0.rhs
    z0 = dsolve(eq,z(x))

    y = y1*z0.rhs
    return y.simplify()
```

## Дифференциальные уравнения

Сначала разберёмся как в питоне функции свои делать. Создаём новую функцию:

```python
def my_fun(x):
    return x ** 2
```

Тестируем новую функцию:

```python
my_fun(5)
```

Система Лотка-Вольтера:

$r$ — численность кроликов (rabbits)

$f$ — численность лис (foxes)

$$
\begin{cases}
r'(t) = r - 0.1 r\cdot f \\
f'(t) = -1.5 f + 0.08 r\cdot f
\end{cases}
$$


Задаём функцию, которая для заданных $r$ и $f$ и момента $t$ находит производные:

```python
def dydt(y, t):
    r, f = y
    return [r - 0.1 * r * f, -1.5 * f + 0.08 * r * f]
```

Тестируем функцию для подсчёта производных:

```python
dydt([10, 7], 5)
```

Начальные условия для дифференциального уравнения:

```python
y0 = [10, 7]
```

Задаём ось времени:

```python
time = np.arange(start=0, stop=10, step=0.5)
```

```python
time
```

Импортируем функцию для решения дифференциального уравнения:

```python
from scipy.integrate import odeint
```

Решаем!

```python
y = odeint(dydt, y0, time)
```

```python
y
```

Выделяем два отдельных вектора: кроликов (r) и лис (f):

```python
r = y[:, 0]
f = y[:, 1]
```

Даёшь график численности:

```python
plt.plot(time, r)
plt.plot(time, f)
```

А теперь приступим к построению фазового портрета. Сначала зададим числа на вертикальной и горизонтальной осях:

```python
r_axis = np.arange(start=0, stop=30, step=3)
f_axis = np.arange(start=0, stop=30, step=3)
```

```python
r_axis
```

Теперь для каждой точки решётки на плоскости посчитаем $r$ и $f$ координату:

```python
all_y = np.meshgrid(r_axis, f_axis)
```

```python
all_r, all_f = all_y
```

```python
all_r
```

```python
all_f
```

Одним махом находим производные во всех этих точках:

```python
dr, df = dydt(all_y, 0)
```

Изображаем:

```python
plt.quiver(all_r, all_f, dr, df)
```

У нас получились стрелочки разной длины. С одной стороны это несёт информацию о скорости изменения переменных, с другой — портит график. Сделаем стрелочки одинаковой длины!

Посчитаем длину стрелочек:

```python
arrow_len = np.sqrt(dr ** 2 + df ** 2)
```

Заменяем нулевую длину на единичную, чтобы избежать проблем с делением на нулевую длину:

```python
arrow_len[arrow_len == 0] = 1
```

Нормируем на длину:

```python
drn = dr / arrow_len
dfn = df / arrow_len
```

Рисуем фазовый портрет:

```python
plt.quiver(all_r, all_f, drn, dfn)
```
